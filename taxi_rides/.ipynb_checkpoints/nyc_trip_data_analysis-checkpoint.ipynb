Next, we can combine the taxi info with the geospatial zone data and make maps. The first map highlights the most popular locations for pickup and drop-off for each for both yellow and green taxis.

```python
# Count yellow taxi pick-ups
# and drop-offs by taxi zone
# normalize for color lookup
yellow_pups = df_yellow['PULocationID'].value_counts()
yellow_pups = 1.0 * yellow_pups / yellow_pups.max()
yellow_doffs = df_yellow['DOLocationID'].value_counts()
yellow_doffs = 1.0 * yellow_doffs / yellow_doffs.max()

# Count green taxi pick-ups
# and drop-offs by taxi zone
# normalize for color lookup
green_pups = df_green['PULocationID'].value_counts()
green_pups = 1.0 * green_pups / green_pups.max()
green_doffs = df_green['DOLocationID'].value_counts()
green_doffs = 1.0 * green_doffs / green_doffs.max()

def build_image(series, shps_nyc, ax, color, title_text):
    """
    Create intensity image 
    against shapefiles
    
    Inputs:
        - series: series of the data
        - shps_nyc: list of map shapes
        - ax: axis handle
        - color: colorbar to sample
        - title_text: text for title of plot
    """
    for row in zip(series, series.index):
        # shp_nyc index
        idx = row[1] - 1

        # ignore the last two
        if idx < 262:
            s = shps_nyc[idx]
            x = [t[0] for t in s.points]
            y = [t[1] for t in s.points]
            ax.fill(x,
                    y,
                    facecolor = color(row[0]),
                    edgecolor='k',
                    linewidth=0.2)
    ax.set_title(title_text)
    ax.axis('off')

# create colorbars
yellows = plt.get_cmap('YlOrRd')
greens = plt.get_cmap('Greens')

# build images
plt.rcParams['figure.figsize'] = (10, 10)
f, axarr = plt.subplots(2, 2)

build_image(yellow_pups, shps_nyc, axarr[0, 0], yellows,'Yellow Cab Pick-Ups - Jan2017')
build_image(yellow_doffs, shps_nyc, axarr[0, 1], yellows, 'Yellow Cab Drop-Offs - Jan2017')
build_image(green_pups, shps_nyc, axarr[1, 0], greens, 'Green Cab Pick-Ups - Jan2017')
build_image(green_doffs, shps_nyc, axarr[1, 1], greens, 'Green Cab Drop-Offs - Jan2017')
plt.savefig('./images/pickups_and_dropoffs_by_region.png')
```

For yellow taxis, pickups mostly take place in Manhattan and at the two airports. The green taxis complement this behavior; since they are not allowed to pickup in parts of Manhattan (Central Park or below) or the airports most of their pickups take place in Harlem and Queens. In both cases it looks like there are large parts of the city where only a few taxi pickups occur.

Drop-offs follow a similar pattern. Some customers are dropped off at the airport, but not as many as the more popular locations in the city.

Taxi wait lot at Laguardia Airport NYC

For the policy analysis section of the work I wanted to focus on airport taxi behavior. As we highlighted above airports are a unique location in NYC for taxis. They are busy and also reliably produce large fares. As a result they can create challenges for maintaining a reliable and efficient system.

The governing body of the NYC airports (not sure if it's city, county or state) should be motivated to maintain an efficient taxiing service at the airports. If there aren't enough taxis then potential passengers will have  to wait in long lines and will (loudly) voice their displeasure. Likewise, since the airport is such a desirable location for taxis to get customers, there is risk of over-suppling taxis and creating logistical headaches for management. Too many taxis at the airport take up space. Since space is such a premium at airports, too many taxis equates to lost revenue. Furthermore, if the car engines are on (for example, in the winter) then the car is also emitting pollutants.

For this analysis we would like to better understand the taxi-passenger queues at LaGuardia airport. We are going to start with a simple model and highlight the assumptions that lead us to it. 

We are also going to start with a small set of data. For now, we will look at the first quarter of 2017, i.e., Jan, Feb and Mar.

```python
def read_and_get_lga_trips(directory, fname):
    """
    read in csvs of taxi data
    and filter down to LGA trips
    """
    # read taxi data
    df = pd.read_csv(directory + fname)
    
    # laguardia is zone 138
    df_lgapups = df.loc[df['PULocationID'] == 138]
    df_lgadoffs = df.loc[df['DOLocationID'] == 138]
    return df_lgapups, df_lgadoffs

# read in 3 months of data
directory = "/Users/dfinkel/proto_dev/data_science/misc/taxi_rides/data/"
yellow_fnames = ["yellow_tripdata_2017-01.csv",
                 "yellow_tripdata_2017-02.csv",
                 "yellow_tripdata_2017-03.csv"]
green_fnames = ["green_tripdata_2017-01.csv",
                "green_tripdata_2017-02.csv",
                "green_tripdata_2017-03.csv"]

df_doffs = []
df_pups = []

# yellow taxi data
for yf in tqdm(yellow_fnames, total=len(yellow_fnames)):
    df1, df2 = read_and_get_lga_trips(directory, yf)
    df_pups.append(df1)
    df_doffs.append(df2)

# green taxi data
for gf in tqdm(green_fnames, total=len(green_fnames)):
    df1, df2 = read_and_get_lga_trips(directory, gf)
    df_pups.append(df1)
    df_doffs.append(df2)

# merge into single dataframes
df_lga_pups = pd.concat(df_pups, axis=0)
df_lga_doffs = pd.concat(df_doffs, axis=0)
```

Some clean-up of the data. Convert the timestamp to seconds after epoch and also extract hour-of-day from the timestamp.

```python
def determine_timestamp(val1, val2):
    """
    determine which val is a nan
    and return the other
    """
    pattern = '%Y-%m-%d %H:%M:%S'
    
    if type(val1) != float:
        return int(time.mktime(time.strptime(val1, pattern)))
    else:
        return int(time.mktime(time.strptime(val2, pattern)))

def determine_hod(val1, val2):
    """
    determine hour of day
    of timestamp
    """    
    if type(val1) != float:
        return int(val1[11:13])
    else:
        return int(val2[11:13])

# time stamp is in different
# columns for green and yellow
# so combine into a single column
# as seconds after epoch
df_lga_doffs['DOtimestamp'] = \
    df_lga_doffs.apply(lambda row: determine_timestamp(row['lpep_dropoff_datetime'],
                                                       row['tpep_dropoff_datetime']), axis=1)

# repeat for pickups
df_lga_pups['PUtimestamp'] = \
    df_lga_pups.apply(lambda row: determine_timestamp(row['lpep_pickup_datetime'],
                                                       row['tpep_pickup_datetime']), axis=1)

# determine hour of day of drop-off
df_lga_doffs['hod'] = df_lga_doffs.apply(lambda row: determine_hod(row['lpep_dropoff_datetime'],
                                                       row['tpep_dropoff_datetime']), axis=1)

# determine hour of day of pick-up
df_lga_pups['hod'] = df_lga_pups.apply(lambda row: determine_hod(row['lpep_pickup_datetime'],
                                                       row['tpep_pickup_datetime']), axis=1)
```

I'd like to model the arrivals of passengers and taxis with a Poission distribution. A core assumption of this approach is that the interarrival times of both systems are distributed exponentially.

As seen in the last section taxi behavior changes as a function of hour-of-day. So I wanted to look at the interarrival times of the system broken out by hour-of-day.

```python
# pick-up analysis
# sort by timestamp
df_lga_pups = df_lga_pups.sort_values('PUtimestamp')

# calc delta seconds between timestamp
df_lga_pups['delta_timestamp'] = df_lga_pups['PUtimestamp'].diff()

# plot distribution by hour
plt.rcParams['figure.figsize'] = (10, 10)
f, axarr = plt.subplots(6, 4)
all_histos = []
for hod in np.arange(24):
    ax = axarr[hod / 4, hod%4]
    data = df_lga_pups.loc[df_lga_pups['hod'] == hod]['delta_timestamp'].fillna(0).values
    h, be = np.histogram(data, bins=np.arange(300))
    ax.plot(be[:-1], 1.0 * h / np.sum(h), color='k')
    ax.set_ylim([0, 0.16])
    ax.text(10, 0.11, 'Hour %i \nMean time = %.2fs' % (hod, np.mean(data)), fontsize=12)
    ax.set_xlabel('Time Between Pickups (sec)', fontsize=10)
f.suptitle('Interarrival Times for Taxi Pickups at LGA Filtered by Hour-of-Day (Jan-Mar 2017)', fontsize=16)
```

```python
# drop-off analysis
# sort by timestamp
df_lga_doffs = df_lga_doffs.sort_values('DOtimestamp')

# calc delta seconds between timestamp
df_lga_doffs['delta_timestamp'] = df_lga_doffs['DOtimestamp'].diff()

# plot distribution by hour
plt.rcParams['figure.figsize'] = (10, 10)
f, axarr = plt.subplots(6, 4)
all_histos = []
for hod in np.arange(24):
    ax = axarr[hod / 4, hod%4]
    data = df_lga_doffs.loc[df_lga_doffs['hod'] == hod]['delta_timestamp'].fillna(0).values
    h, be = np.histogram(data, bins=np.arange(300))
    ax.plot(be[:-1], 1.0 * h / np.sum(h), color='k')
    ax.set_ylim([0, 0.16])
    ax.text(10, 0.11, 'Hour %i \nMean time = %.2fs' % (hod, np.mean(data)), fontsize=12)
    if hod >= 20:
        ax.set_xlabel('Time Between Drop-offs (sec)', fontsize=10)
f.suptitle('Interarrival Times for Taxi Drop-offs \
            at LGA Filtered by Hour-of-Day (Jan-Mar 2017)', fontsize=16)
```

You can see that during non-peak hours the systems break down and the interarrival times between drop-off / pick-ups are messy. However, in peak hours the distributions appear exponential.

We can look at this more carefully by comparing the curves to the exponential distribution. Between 2 - 4pm the data appears consistent for pickups and drop-offs, respectively. Let's compare the data with theoretical exponential distributions.

```python
# compare interarrival distribution
# to theory and demonstrate 
# drop-offs and pickups are Poisson Process
plt.rcParams['figure.figsize'] = (10, 5)
f, axarr = plt.subplots(1, 2)

# look at pickup data between 2-4pm
pup_data = df_lga_pups.loc[df_lga_pups['hod'] \
                      .isin([14, 15])]['delta_timestamp'] \
                      .fillna(0).values

# Calc distribution of pickup data
h, be = np.histogram(pup_data, bins=np.arange(100))

# plot distribution of pickup data
axarr[0].plot(be[:-1], 1.0 * h / np.sum(h), color='k', linewidth=2)
axarr[0].set_ylim([0, 0.12])
axarr[0].text(10, 0.09, 'Hour 2-4pm \nMean time = %.2fs' % np.mean(pup_data), fontsize=12)
axarr[0].set_xlabel('Time Between Pickups (sec)')
axarr[0].set_ylabel('Frequency')
axarr[0].set_title('Demonstrating Poisson Fit for LGA Pickups')

# fit exponential distribution to pickup data
x = np.arange(100)
loc, scale = scp.stats.expon.fit(pup_data)
y = scp.stats.expon.pdf(x, loc, scale)
axarr[0].plot(x, y, '--', color='darkred', linewidth=2)
axarr[0].legend(['Data', 'Theory'])

# look at dropoff data between 2-4pm
doff_data = df_lga_doffs.loc[df_lga_doffs['hod'] \
                        .isin([14, 15])]['delta_timestamp'] \
                        .fillna(0).values

# Calc distribution of drop-off data
h, be = np.histogram(doff_data, bins=np.arange(100))

# plot distribution of dropoff data
axarr[1].plot(be[:-1], 1.0 * h / np.sum(h), color='k', linewidth=2)
axarr[1].set_ylim([0, 0.12])
axarr[1].text(10, 0.07, 'Hour 2-4pm \nMean time = %.2fs' % np.mean(doff_data), fontsize=12)
axarr[1].set_xlabel('Time Between Drop-offs (sec)')
axarr[1].set_ylabel('Frequency')
axarr[1].set_title('Demonstrating Poisson Fit for LGA Drop-offs')

# fit exponential distribution to dropoff data
x = np.arange(100)
loc, scale = scp.stats.expon.fit(doff_data)
y = scp.stats.expon.pdf(x, loc, scale)
axarr[1].plot(x, y, '--', color='darkred', linewidth=2)
axarr[1].legend(['Data', 'Theory'])
plt.savefig('./data_vs_poisson_theory.png')
```

The fit appears really strong. The maroon dashed line lines up really well the curves computed from the data. I am going to use this as evidence to validate our assumption and proceed with Poisson distribution models for the arrivals of taxis and passengers.

Note that the Poisson Distribution is:
$$P \left(N(t)=n \right) = \frac{(\lambda t)^n}{n!} \exp^{- \lambda t} $$

For this analysis I am making the following assumptions:
1. Taxis arrive at the airport according to a Poisson Process
2. Passengers arrive for taxi pickup according to a Poisson Process
3. Taxis that drop-off a passenger immediately enter a "virtual queue" for passenger pickup
4. The virtual queue can be thought of as a phone-based app modeled after Uber/Lyft. A passenger requests a taxi and the app enters them into a queue. As soon as a taxi is available the app "assigns" the available taxi to the passenger. Even though physical pickup still needs to occur, we consider the assignment complete.
5. When a taxi arrives at the airport and there are no passengers waiting it enters a FIFO queue, like the photo shown above.
6. We assume the airport is a single location. In real-life, taxis pick-up and drop-off at different terminals

To start I was curious to know if the taxi traffic created by drop-offs was enough to handle all of the pickups that take place at the airport. To do this we can look at the distribution of arrivals and departures from our model.

```python
"""
How many taxi pickups typically
occur over two hour period and
compare to drop-offs
"""
# seconds
t = 120*60

lam_pups = 1./np.mean(pup_data)
lam_doffs = 1./np.mean(doff_data)
poisson_pups = []
poisson_doffs = []
for n in np.arange(0, 1000):
    poisson_pups.append(scp.stats.poisson.pmf(n, t*lam_pups))
    poisson_doffs.append(scp.stats.poisson.pmf(n, t*lam_doffs))
    
plt.plot(np.arange(0, 1000), poisson_pups)
plt.plot(np.arange(0, 1000), poisson_doffs)
plt.xlabel('Number of Events (i.e., Drop-offs or Pickups)')
plt.ylabel('Frequency')
plt.title('Distribution of Drop-offs and Pickups')
plt.legend(['pickups', 'drop-offs'])
```

The distributions shown above illustrate that drop-offs alone do not create enough taxis needed for all of the pickups that take place at the airport. The distributons are disjoint and there is effectively 0 probability that drop-offs will ever create enough taxis to meet pickup demand (once our simulation is up and running we can look at this case and demonstrate how long a passenger would wait if only drop-off taxis could pick-up passengers).

Instead, additional taxis must arrive at the airport to meet the demand of passengers. If we assume those taxis arrive at a Poisson rate then our overall arrival rate of taxis can still be modeled as a Poisson Process. That is,if $X_i \sim P(\lambda_i), i = 1, ..., n$ are independent and $\lambda = \sum_i \lambda_i$, then $Y = \sum_i X_i \sim P(\lambda)$. We will parameterize the rate of extra taxis arriving at the airport and simulate the impact of those taxis. If too many taxis arrive then the taxi queue will get too big and the taxi parking lot will fill up. Conversely, if too few extra taxis show up at the airport then the passengers will have to wait a long time for their ride.

Although there are closed-form equations for this queue model I decided to create a simple simulation that we can use to evaluate queue behavior. I chose this path because additional complexity can be easily added. I also wanted to explore different aspects of the queue behavior (such as max taxi queue) over a finite timeline -- a Monte Carlo simulation seemed appropriate to perform this analysis.

In the code below I build the simulation and then run it on the scenario when there are no extra taxis deployed to the airport, i.e., only taxis that drop passengers off are eligible to pick passengers up.

```python
def airport_taxi_sim(lam_pups, lam_doffs, lam_extra):
    """
    simulate the Poisson process
    of taxi pickups at airport
    
    Inputs:
        lam_pups : rate parameter for pickups
        lam_doffs: rate parameter for dropoffs
        lam_extra: rate parameter for extra taxis to help
        
    Outputs:
        pickup_times: pickup events on two-hour clock
        passenger_wait_times: wait time for each passenger to get a taxi
        taxi_wait_times: wait time for each taxi to get a passenger
        taxi_queue_len: len of taxi queue (referenced to each pickup event)
        
    Assumptions:
        1. airport treated as a single queue
        2. drop-off taxis immediately enter queue
        3. pickups occur virturally, i.e., when taxi gets assigned
           a passenger
        4. system is closed, i.e., no one gets fed up and leaves
    """
    
    # simulate interarrival times
    pickup_inters = np.random.exponential(1./lam_pups, 1100)
    # convert to wall clock events
    pickup_times = np.cumsum(pickup_inters)
    # filter to only customers that arrive in the 2 hour window
    pickup_times = pickup_times[pickup_times < (120 * 60)]
    
    # adjust drop-off rate parameter
    # for extra taxis that come for fares
    lam_doffs += lam_extra

    # simulate interarrival times
    dropoff_inters = np.random.exponential(1./lam_doffs , 1100)
    # convert to wall clock events
    dropoff_times = np.cumsum(dropoff_inters)
    # only get enough taxis to pickup each passenger
    dropoff_times = dropoff_times[0:len(pickup_times)]

    # calc mean wait time for passengers
    # calc mean wait time for taxi
    # calc avg queue length for taxi
    passenger_wait_times = []
    taxi_wait_times = []
    taxi_queue_len = []
    for ii in np.arange(len(pickup_times)):
        delta_time = dropoff_times[ii] - pickup_times[ii]
        if delta_time > 0:
            # passenger had to wait
            passenger_wait_times.append(delta_time/60.)
            taxi_wait_times.append(0)
            taxi_queue_len.append(0)
        else:
            # taxi had to wait
            passenger_wait_times.append(0)
            taxi_wait_times.append(-1*delta_time/60.)

            # how long did queue get?
            tmp = dropoff_times[dropoff_times < pickup_times[ii]]
            taxi_queue_len.append(len(tmp[[tmp >= dropoff_times[ii]]]))
    return pickup_times, passenger_wait_times, taxi_wait_times, taxi_queue_len

# pickup rate
lam_pups = 1./np.mean(pup_data)

# drop-off rate
# assume every taxi that drops-off
# immediately enters the queue for pickups
lam_doffs = 1./np.mean(doff_data)

# extra taxis that come
# to airport without a drop-off
lam_extra = 0.

# single MC simulation
pickup_times, passenger_wait_times, taxi_wait_times, taxi_queue_len = \
    airport_taxi_sim(lam_pups, lam_doffs, lam_extra)

# plot single MC simulation
plt.rcParams['figure.figsize'] = (10, 5)
f, axarr = plt.subplots(1, 2)
axarr[0].plot(pickup_times/60., passenger_wait_times)
axarr[0].plot(pickup_times/60., taxi_wait_times)
axarr[0].legend(['passenger', 'taxi'])
axarr[0].set_xlabel('Clock Time (minutes after 2pm EST)')
axarr[0].set_ylabel('Wait time (minutes)')
axarr[0].set_title('Passenger and Taxi Wait Times \nNo Extra Taxis')

axarr[1].plot(pickup_times/60., taxi_queue_len, 'k')
axarr[1].set_xlabel('Clock Time (minutes after 2pm EST)')
axarr[1].set_ylabel('Wait time (minutes)')
axarr[1].set_title('Taxi Queue Length \nNo Extra Taxis')
```

As expected, the taxis never queue while the passengers end up in some pretty long lines! The passengers that arrive at the end of the time period, i.e., 4pm, are looking at almost a 50 minute wait for their taxi. Also, aside from some noise at the start there is no wait and no queue formed for the taxis. They are in very high demand in this scenario.

Lets run the sim again and add some extra taxis to the scenario. We'll assume that an extra taxi arrives according to a Poisson Process with mean arrival time of every 30 seconds.

```python
# extra taxis that come
# to airport without a drop-off
lam_extra = 1./30

# single MC simulation
pickup_times, passenger_wait_times, taxi_wait_times, taxi_queue_len = \
    airport_taxi_sim(lam_pups, lam_doffs, lam_extra)

# plot single MC simulation
plt.rcParams['figure.figsize'] = (10, 5)
f, axarr = plt.subplots(1, 2)
axarr[0].plot(pickup_times/60., passenger_wait_times)
axarr[0].plot(pickup_times/60., taxi_wait_times)
axarr[0].legend(['passenger', 'taxi'])
axarr[0].set_xlabel('Clock Time (minutes after 2pm EST)')
axarr[0].set_ylabel('Wait time (minutes)')
axarr[0].set_title('Passenger and Taxi Wait Times \nSome Extra Taxis')

axarr[1].plot(pickup_times/60., taxi_queue_len, 'k')
axarr[1].set_xlabel('Clock Time (minutes after 2pm EST)')
axarr[1].set_ylabel('Wait time (minutes)')
axarr[1].set_title('Taxi Queue Length \nSome Extra Taxis')
plt.savefig('Queue_analysis.png')
```

The plots show this is a much more desirable scenario for passengers. In our simulation, passengers never waited more than a few minutes in line and in most cases they were assigned a taxi immediately. Taxis did queue up over the timeline -- at one point there is a 55 taxi queue that the airport will need to support. However, the wait time for a taxi in the queue isn't too bad, peaking at around 7 minutes.

Lets finish by parameterizing the extra taxis that arrive at the airport and characterize the impact on passenger waits, taxi waits and taxi queue lengths.

```python
"""
MC Simulate
Vary number of taxis that come
to the airport to support pickups
"""

# how many monte carlos?
mc = 100

lam_extras = [1./60, 1./50, 1./40, 1./30, 1./20, 1./10, 1./5]
all_taxi_queue = []
all_taxi_wait = []
all_pass_wait = []
for lam_extra in tqdm(lam_extras, total=len(lam_extras)):    
    pass_wait = []
    taxi_wait = []
    taxi_queue = []
    for ii in np.arange(mc):
        pickup_times, passenger_wait_times, taxi_wait_times, taxi_queue_len = \
            airport_taxi_sim(lam_pups, lam_doffs, lam_extra)
        pass_wait.append(np.max(passenger_wait_times))
        taxi_wait.append(np.max(taxi_wait_times))
        taxi_queue.append(np.max(taxi_queue_len))
    all_pass_wait.append(np.mean(pass_wait))
    all_taxi_wait.append(np.mean(taxi_wait))
    all_taxi_queue.append(np.mean(taxi_queue))
```

```python
# Plot simulation results
fig, ax1 = plt.subplots()
ax1.plot(all_taxi_wait, all_pass_wait, 'darkblue', linewidth=3)
ax1.set_xlabel('Taxi Wait Time (minutes)')
ax1.set_ylabel('Passenger Wait Time (minutes)', color='darkblue')
ax1.tick_params('y', colors='darkblue')

ax2 = ax1.twinx()
ax2.plot(all_taxi_wait, all_taxi_queue, 'darkred')
ax2.set_ylabel('Taxi Queue Length', color='darkred')
ax2.tick_params('y', colors='darkred')

fig.tight_layout()
plt.title('Taxi Queue vs Passenger Wait')
plt.savefig('./images/queue_vs_wait.png')
```

Although very preliminary, these curves suggest that lot is unneccessarily big to handle passenger pickup traffic in the 2-4pm window. If that lot is full taxis will need to wait almost 60 minutes for a passenger. If that lot was only half full then airport passengers would not see much degradation in their queue length. The airport would have fewer taxis on the property and space would be freed for other airport operations.

This model is a very simple model with many assumptions. Next steps would be to begin peeling back these assumptions such as adding additional lags to the system to account for transport to/from the taxi lot. A more realistic passenger model could also be added that could be a function of actual aircraft arrival traffic. We could also add distribution models to the rate parameter and switch a more sophisticated MCMC simulation.

If additional attributes are available from the taxis the model could also be improved. For example, if the airport was broken up into zones based on terminals then the single queue assumption could be improved. It would also be valuable to request car ID's for each trip. This would help refine the assumptions about how many taxis that drop-off passengers stick around for a pick-up as well as how many taxis arrive without a passenger to drop-off.
